```python
search_product.send_keys("Laptop")
```

```python
filter_button=driver.find_elements_by_xpath("//a[@class='a-link-normal s-navigation-item']/span")
for i in filter_button:
    if i.text == 'Intel Core i7':
        i.click()
        break
```

```python
urls = []
for i in range(0,3):
    for j in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']/a"):
        urls.append(j.get_attribute("href"))
    driver.find_element_by_xpath('//li[@class="a-disabled"]').click()
    time.sleep(4)
```

```python
title=[]
rating=[]
price =[]
for i in range(0,10):
    for j in driver.find_elements_by_xpath("//h2[@class='a-size-mini a-spacing-none a-color-base s-line-clamp-2']"):
        title.append(j.text)
    try:
        for k in driver.find_elements_by_xpath("//i[@class='a-icon a-icon-star-small a-star-small-4 aok-align-bottom]"):
            rating.append(k.text)
    except NoSuchElementException:
        rating.append('-')
    try:
        for l in driver.find_elements_by_xpath("//span[@class='a-price']"):
            price.append(l.text)
    except NoSuchElementException:
            rating.append('-')
    driver.find_element_by_xpath('//li[@class="a-disabled"]').click()
    time.sleep(4)
```
